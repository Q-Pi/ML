```python
df = pd.read_csv("data/owid-covid-data.csv")
df2 = pd.read_csv("data/EDU_DEM_12092020154828100.csv")
```

```python
df2 = df2.query("YEAR == 2018 & SEX == \"T\" & AGE == \"T\"")
```

```python
merged = df.merge(df2, left_on='location', right_on='Country', how='left')
```

```python
#merged.fillna(0).query("Value != 0")
```

```python
merged['infected_percent'] = 100*merged['total_cases']/merged['Value']
```

```python
info1 = merged[['location', 'infected_percent', 'total_cases', 'Value']].fillna(0).query('infected_percent != 0').groupby(by='location').max().sort_values('infected_percent', ascending=0)
```

```python
dateMax = df.query('location == \"France\"')['date'].max()
sss = dateMax.split('-')
sss.reverse()
dateMax = "/".join(sss)
```

```python
p = info1.query("location == \"France\"")["infected_percent"][0]
print("Pourcentage de personne infecté par le COVID-19 en France au", dateMax, ":", '{:02.2f}'.format(p), "%")
```
