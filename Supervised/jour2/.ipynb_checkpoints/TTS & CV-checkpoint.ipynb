### Unnamed

```python
import seaborn as sns; sns.set()

#Chargement du dataset
iris = sns.load_dataset('iris')

iris.shape
```

```python
#Pairplot
sns.pairplot(iris, hue='species', corner=True, height=2)
```

```python
#Isolation
Y = iris["species"]
X = iris.drop(columns='species')
```

```python
from sklearn.model_selection import train_test_split

#Préparation pour TTS
splits = train_test_split(X, Y, 
                         train_size = 0.3,
                         stratify=Y,
                         random_state=6)

#for each in splits:
#    print(each.shape)

X_train, X_test, Y_train, Y_test = splits
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN #KNeighborsRegressor
knn = KNN(n_neighbors=3)

#Entrainement
knn.fit(X_train, Y_train)
```

```python
#Test
predictions_test = knn.predict(X_test)
```

```python
#Créer une mesure de performance
def accuracy(preds, Y):
    return (preds == Y).sum() / len(Y)

#Performance du test
accuracy(predictions_test, Y_test)
```

## Cross Validation

```python
knn = KNN(n_neighbors=3)
from sklearn.model_selection import cross_val_score
result = cross_val_score(knn, X_train, Y_train, cv=10, scoring='accuracy')
sum(result)/len(result)
```
